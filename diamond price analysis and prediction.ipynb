# dealing with outliers

```python
df.columns
```

```python
q3= df.carat.quantile(0.75)
q1= df.carat.quantile(0.25)
iqr = q3-q1
lower_threshold = q1-1.5*iqr
upper_threshold = q3+1.5*iqr
print("upper_threshold--",upper_threshold)
print("lower_threshold--",lower_threshold)
df.carat= df.carat.clip(lower_threshold,upper_threshold)
```

```python
#depth
q3= df.depth.quantile(0.75)
q1= df.depth.quantile(0.25)
iqr = q3-q1
lower_threshold = q1-1.5*iqr
upper_threshold = q3+1.5*iqr
print("upper_threshold--",upper_threshold)
print("lower_threshold--",lower_threshold)
df.depth= df.depth.clip(lower_threshold,upper_threshold)
```

```python
#table
q3= df.table.quantile(0.75)
q1= df.table.quantile(0.25)
iqr = q3-q1
lower_threshold = q1-1.5*iqr
upper_threshold = q3+1.5*iqr
print("upper_threshold--",upper_threshold)
print("lower_threshold--",lower_threshold)
df.table= df.table.clip(lower_threshold,upper_threshold)
```

```python
#price
q3= df.price.quantile(0.75)
q1= df.price.quantile(0.25)
iqr = q3-q1
lower_threshold = q1-1.5*iqr
upper_threshold = q3+1.5*iqr
print("upper_threshold--",upper_threshold)
print("lower_threshold--",lower_threshold)
df.price= df.price.clip(lower_threshold,upper_threshold)
```

```python
#size
q3= df.Size.quantile(0.75)
q1= df.Size.quantile(0.25)
iqr = q3-q1
lower_threshold = q1-1.5*iqr
upper_threshold = q3+1.5*iqr
print("upper_threshold--",upper_threshold)
print("lower_threshold--",lower_threshold)
df.Size= df.Size.clip(lower_threshold,upper_threshold)
```

# fit the data in variables

```python
X = df[['carat',  'Size', 'color_D', 'color_E',
       'color_F', 'color_G', 'color_H', 'color_I', 'color_J', 'cut_Fair',
       'cut_Good', 'cut_Ideal', 'cut_Premium', 'cut_Very Good', 'clarity_I1',
       'clarity_IF', 'clarity_SI1', 'clarity_SI2', 'clarity_VS1',
       'clarity_VS2', 'clarity_VVS1', 'clarity_VVS2']].values
```

```python
y = df["price"].values
```

# spliiting the data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# scaling the data

```python
from sklearn.preprocessing import StandardScaler ## standrard scalig 
scaler = StandardScaler() #initialise to a variable
scaler.fit(X_train) # we are finding the values of mean and sd from the td
X_train = scaler.transform(X_train) # fit (mean, sd) and then transform the training data
X_test = scaler.transform(X_test) # transform the test data 
```

# linear regression

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

```python
dflr = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
dflr.head()
```

```python
regressor.intercept_ # c 
```

```python
regressor.coef_
```

```python
from sklearn import metrics
print('R2- SCORE:', metrics.r2_score(y_test,y_pred))
```

# Knn regressor

```python
X = df[['carat', 'depth', 'table',  'Size', 'color_D', 'color_E',
       'color_F', 'color_G', 'color_H', 'color_I', 'color_J', 'cut_Fair',
       'cut_Good', 'cut_Ideal', 'cut_Premium', 'cut_Very Good', 'clarity_I1',
       'clarity_IF', 'clarity_SI1', 'clarity_SI2', 'clarity_VS1',
       'clarity_VS2', 'clarity_VVS1', 'clarity_VVS2']].values
y = df["price"].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler ## standrard scalig 
scaler = StandardScaler() #initialise to a variable
scaler.fit(X_train) # we are finding the values of mean and sd from the td
X_train = scaler.transform(X_train) # fit (mean, sd) and then transform the training data
X_test = scaler.transform(X_test) # transform the test data 
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score
```

for i in range(1,15):
  knn_regressor=KNeighborsRegressor(n_neighbors=i)
  knn_regressor.fit(X_train,y_train)
  print("K value : ", i  ,"cross_val_score : ", cross_val_score(knn_regressor,X_train,y_train,cv = 10).mean())

```python
#best k value is 3
knn_regressor=KNeighborsRegressor(3)
knn_regressor.fit(X_train,y_train)
```

```python
y_pred = knn_regressor.predict(X_test)
```

```python
dfknn = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
dfknn.head()
```

```python
from sklearn import metrics
print('R2- SCORE:', metrics.r2_score(y_test,y_pred))
```

# xg boost

```python
import xgboost as xgb
```

```python
for lr in [0.01,0.05,0.08,0.09,0.1,0.11,0.12,0.13,0.15,0.18,0.2,0.25,0.3]:
  xgb_regressor= xgb.XGBRegressor(colsample_bytree=0.4,learning_rate = lr,n_estimators=1000) 
  xgb_regressor.fit(X_train,y_train) 
  print("Learning rate : ", lr,"cross_val_score:", cross_val_score(xgb_regressor,X_train,y_train,cv = 15).mean())
```

```python
#best learning rate is 0.08
```

```python
xgb_regressor= xgb.XGBRegressor(learning_rate = 0.08,n_estimators=1000) 
xgb_regressor.fit(X_train,y_train) 
```

```python
y_pred = xgb_regressor.predict(X_test)
```

```python
dfxgb = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
dfxgb.head()
```
